## Check POI in map(Korea)

```python
import sys, os
import numpy as np
import pandas as pd
import pyproj
import folium
```

```python
POI_data = pd.read_csv("POI/문화/관광숙박업.csv", encoding='cp949', usecols=['위도', '경도'])
```

```python
print(POI_data)
```

```python
# 지도 중심 좌표 설정
lat_c, lon_c = 37.53165351203043, 126.9974246490573

# Folium 지도 객체 생성
m = folium.Map(location=[lat_c, lon_c], zoom_start=6)

# 마커 생성
for _, row in POI_data.iterrows():
    lat, lon = row['위도'], row['경도']
    folium.Marker(location=[lat, lon]).add_to(m)
    
m
```
